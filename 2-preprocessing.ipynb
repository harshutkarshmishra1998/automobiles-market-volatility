1 - Import Libraries

```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

2 - Load Dataset

```python
raw_data = pd.read_csv("automobile-indexes.csv", index_col=["Date"])
raw_data.index = pd.to_datetime(raw_data.index)
raw_data.head()
```

```python
df = pd.DataFrame([])

# Market Index
df['vol'] = raw_data['VW']
df['por'] = raw_data['POR']
df['bmw'] = raw_data['BMW']

# Creating Returns
df['ret_vol'] = raw_data['VW'].pct_change(1).mul(100)
df['ret_por'] = raw_data['POR'].pct_change(1).mul(100)
df['ret_bmw'] = raw_data['BMW'].pct_change(1).mul(100)

# Creating Squared Returns
df['sq_vol'] = df.ret_vol.mul(df.ret_vol)
df['sq_por'] = df.ret_por.mul(df.ret_por)
df['sq_bmw'] = df.ret_bmw.mul(df.ret_bmw)

# Volume
df['q_vol'] = raw_data['Vol_VW']
df['q_por'] = raw_data['Vol_POR']
df['q_bmw'] = raw_data['Vol_BMW']

df.head()
```

```python
df = df.iloc[1:]
df.head()
```
